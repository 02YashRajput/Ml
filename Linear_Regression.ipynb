```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([10,20,30,40,50,60,70])
y = np.array([100,120,140,160,180,200,220])
w = 0
b = 0
a = 0.0001
m = len(x)
iter = 1000
for i in range(iter):
    y_pred = w*x+b
    error =(y_pred-y)
    w =w- ((a/m)*np.sum((error*x)))
    b = b- ((a/m)*np.sum(error))

plt.plot(x, y, 'o', label='Data points')
plt.plot(x, (w * x + b), label='Regression line')
plt.xlabel('Sq. ft. area')
plt.ylabel('Price')
plt.title('Gradient Descent for Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

x = np.array([[10, 12],
              [20, 10],
              [30, 8],
              [40, 6],
              [50, 4],
              [60, 2],
              [70, 0]])
y = np.array([[100], [120], [140], [160], [180], [200], [220]])
w = np.array([[0], [0]])
b = 0
a = 0.0001
m = len(y)  
iter = 5000

for i in range(iter):
    y_pred = np.dot(x, w) + b 
    error = y_pred - y
    w =w- (a / m) * np.dot(x.T, error)
    b = b- (a / m) * np.sum(error)
plt.scatter(x[:, 0], y, color='blue', label='Data Points')
plt.plot(x[:, 0], y_pred, color='red', label='Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression using Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```
